```python
import yfinance as yf
import polars as pl
```

```python
ibov = pl.from_pandas(yf.download("^BVSP", start="2020-01-01", end="2026-01-01", progress=False).reset_index())
ibov = ibov.rename({col:col.split("'")[1] for col in ibov.columns})
```

```python
ibov = ibov.with_columns(
    (
        (pl.col("Close") / pl.col("Close").shift(1) - 1)
        .alias("return")
    )
)
ibov
```
